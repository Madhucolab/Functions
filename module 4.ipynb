```python
#1.What is the difference between a function and method in python
(method)
Method definitions are always present inside a class
Associated with the class object.
It is called on an object.
It depends on the class they belong to.
It requires the self as its first argument
def add(x, y):
  return x+y result = add(10, 9)
result
output:19

(Function)
No class is needed to define a function
Not associated with any objects.
It is called by its name.
It doesn’t require any self-argument.
it doesn’t depend on any class,it is an identical entity.
def greet(name):
    return f"barela, {name}!"

print(greet("jabalpur"))  
Output: barela,jabalpur! 

#2.Explain the concept of funtion arguments and parameters in python?
Ans:-Arguments An argument is a value that is passed to a function when it is
called. It might be a variable, value or object passed to a function or method as input.
Arguments are the actual values that you pass to the function when you call it. These
values replace the parameters defined in the function.
example:
def add(a, b):
    return a + b
# Here the values 5,3 are arguments
print(add(5, 3))  
Output: 8

(parameters)
Parameters are variables defined in a function declaration. This act as placeholders for the
values (arguments) that will be passed to the function.
# Here a,b are the parameters
def sum(a,b):
  print(a+b)
  
sum(1,2)
output:-3

#3.What are the different ways to define and call a function in python.
Ans:-standard Function Definition
Function with Default Parameters
Function with Variable-Length Arguments
Anonymous Functions
Use the def keyword
Write the function name
Add parentheses and a colon
Add parentheses and a colon
Indent with a tab or four space

#4.What is the purpose of the 'return' statement in a python funtion?
Ans:-A return statement is used to end the execution of the function call and it “returns” 
the value of the expression following the return keyword to the caller. 
The statements after the return statements are not executed. If the return statement is without any expression, 
then the special value None is returned. A return statement is overall used to invoke a function so that the passed statements can be executed.

def square(num):
    return num * num

result = square(4)  # result is now 16
print(result)  
Output: 16

#5.What are iterators in python and how do they differ from iterables.
Ans:-An iterator in Python is an object that holds a sequence of values and provide sequential traversal through a collection of items such as lists,
tuples and dictionaries.The Python iterators object is initialized using the iter() method. It uses the next() method for iteration..
An Iterable  is basically an object that any user can iterate over. 
An Iterator is also an object that helps a user in iterating over another object (that is iterable). 
We can generate an iterator when we pass the object to the iter() method.
    
# list of cities
cities = ["mp", "bihar", "up"]

# initialize the object
iterator_obj = iter(cities)
print(next(iterator_obj))
print(next(iterator_obj))
print(next(iterator_obj))

#6.Explain the concept of generators in python and how they are defined.
Ans:- A generator function is a special type of function that returns an iterator
object. Instead of using return to send back a single value, generator functions use
yield to produce a series of results over time. This allows the function to generate
values and pause its execution after each yield, maintaining its state between
iterations.
 example:
def fun(max):
    cnt = 1
    while cnt <= max:
        yield cnt
        cnt += 1
        
ctr = fun(5)
for n in ctr:
    print(n)
output:-1
        2
        3
        4
        5
explain:-This generator function fun yields numbers from 1 up to a
specified max. Each call to next() on the generator object resumes execution right after the
yield statement, where it last left off.

#7.What are the advantages of using generators over regular functions.
Ans:-1. Efficient Memory usage: Generators do not store the entire sequence in
memory, making them suitable for large datasets or infinite sequence.

2. Simpler syntex: Using yield simplifies the creation of iterators, making the code
more concise and readble.

3. speed: Generators can be faster than regular loops because they avoid the
overhead of creating a list or other data structure to store the results.

4. lazy evaluation: Generators can produce a sequence of values on the fly, which
allows for lazy evaluation.
    
5. flexibiity: Generators can generate data dynamically, allowing for
more flexibility in your code.
example:
def countdown(n):
  while n>0:
    yield n
    n -=1
 #Creating a generators
gen = countdown(5)
 # using the generators
 for num in gen:
   print(num)
     
#8.What is a lambda function in python and when is it typically used?
Ans:-A lambda function in Python is a small, anonymous function defined using the lambda keyword. 
Unlike regular functions defined with def, lambda functions are written in a single line and are often used for short,
simple operations.
Syntax of a Lambda Function:lambda arguments: expression
example:-
square = lambda x: x*x print(square(10))

#9.Explain the purpose and usage of the 'map()','reduce()',and 'filter()' functions in python?
map()
The map() function applies a given function to each item in an iterable (list, tuple) and returns a map object, which is an iterator.

Syntax:filter(function, iterable)
# Function to return double of n
def double(n):
    return n * 2

# Using map to double all numbers
numbers = [5, 6, 7, 8]
result = map(double, numbers)
print(list(result))
output:-10
        12
        14
        16

reduce()
The reduce function is used to apply a particular function passed in its argument to all of the list elements mentioned in the sequence passed along.
This function is defined in “functools” module.
    
syntax:-from functools import reduce
reduce(function, iterable[, initializer])

from functools import reduce
# Find the product of all numbers in the list
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  
Output: 24

filter()
The filter() method filters the given sequence with the help of a function that tests each element in the sequence to be true or false.

Syntax:-filter(function, iterable).

# Filter out even numbers from the list
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  
Output: [2, 4, 6]

#10.What is the differene between 'map()','reduce()',and 'filter()' functions in python? 
map() 
Map applies a function to each element in an array and returns an
array of the results.
example:
s = ['1', '2', '3', '4'] res = map(int, s) print(list(res))

reduce()
Reduce applies a function to the array and returns a single output.
 example:
 from functools import reduce
 # Summing numbers with reduce and lambda
 a = [1, 2, 3, 4, 5]
 res = reduce(lambda x, y: x + y, a)
 print(res)

filter()
Filter uses a boolean function to select certain elements from an array.
example:
 a = [1, 2, 3, 4, 5, 6]
 b = filter(lambda x: x % 2 == 0, a)
 print(list(b))

#11.Using pen & paper write the internal machanism for sum operator using reduce function on this given list:[47,11,42,13]; 
To understand the internal mechanism of the reduce() function for computing the sum of a list, 
let’s break it down step by step for the given list [47, 11, 42, 13].

We’ll use the reduce() function from Python’s functools module.function lambda x, y: x + y
case1:- x = 47, y = 11
result = x + y = 47 + 11 = 58

case2:- x = 58, y = 42
result = x + y = 58 + 42 = 100

case3:- 

```

```python
#1.Write a python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def even_sum(list):
    sum =0
    for num in list:
        if num%2==0:
            sum =sum+num
    return sum
```

```python
#4. 4.Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):

    if num <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True


# Testing the function for numbers 1 to 200
for n in range(1, 201):
    print(f"{n} is {'prime' if is_prime(n) else 'not prime'}")
```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci(n):
    l = [0,1]
    for i in range(2,n):
        l.append(l[-1]+l[-2])
    return l
#Main function
if __name__ == "__main__":
#defining the total number of elements
    n = 11
#calling of function
fibo = fibonacci(n)
#displaying the function
print("Fibonacci Series: ",*fibo)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
for power in powers_of_two(5):
    print(power)
```

```python
# 7.Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filepath):
    try:
       with open(filepath, 'r') as file:
           for line in file:
                yield line.strip() #Yield each after removing leading/trailing whitespace
    except FileNotFoundError:
       print(f"Error: File '{filepath}' not found.")
       yield None  # Yield None to indicate an error
```

```python
#9.Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
#Converts Celsius to Fahrenheit.
  return (celsius * 9/5) + 32
# Example usage:
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))
fahrenheit_temps
```

```python
#10.Create a Python program that uses filter() to remove all the vowels from a given string.
def remove_vowels(input_string):
  vowels = "aeiouAEIOU"
  return "".join(filter(lambda char: char not in vowels, input_string))
# Example usage
string = "mangalayatan"
string_without_vowels = remove_vowels(string)
print(string_without_vowels) 
```
